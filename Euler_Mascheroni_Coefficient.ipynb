Evaluating the convergence value of the Expectation error:

```python
import numpy as np
import matplotlib.pyplot as plt
n=np.array(np.arange(100,5000))
def Error(n):
    error=[]   
    for i in range(0,len(n)):
        temp=[]  # a dummy list for the error in each values of n
        for r in range(1,n[i]+1):
            temp.append(np.ceil(n[i]/r)-(n[i]/r))
        error.append(temp)
    return(error)
def E_of_error(error): #Expectation of error
    E=[]
    for i in range(0,len(error)):
        E.append(sum(error[i])/len(error[i]))
    return E
plt.plot(n,E_of_error(Error(n)),color='orange')
plt.ylim(0.5,0.625)
plt.xlabel("n")
plt.ylabel("Expectation value")
plt.title("Convergance Value of Expectation Error")
print("The value of convergence:",(sum(E_of_error(Error(n)))/len(n)))
```
